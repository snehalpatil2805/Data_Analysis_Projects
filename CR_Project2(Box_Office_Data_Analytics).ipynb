## Import Libraries

```python
import numpy as np
import pandas as pd 
pd.set_option('max_columns',None)
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
plt.style.use('ggplot')
```

## Task 1: Data Loading and Exploration

```python
df=pd.read_csv('box_office.csv')
```

```python
df.head()
```

## Task 2: Visualizing the Target Distribution

```python
fig,ax=plt.subplots(figsize=(16,6))
plt.subplot(1,2,1)
sns.distplot(df['revenue'],kde=False)
plt.title('Distribution of Revenue')
plt.subplot(1,2,2)
sns.distplot(np.log1p(df['revenue']),kde=False)
plt.title('Distribution of Log=Transformed Revenue')
```

## Task 3: Relationship between Film Revenue and Budget

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.scatterplot(df['budget'],df['revenue'])
plt.title('Revenue vs Budget');
plt.subplot(1,2,2)
sns.scatterplot(np.log1p(df['budget']),df['revenue'])
plt.title('Log Revenue vs Log Budget')
```

## Task 4: Distribution of Languages in Film

```python
df.loc[df['original_language'].isin(df['original_language'].value_counts().head(10).index)]
```

```python
language_data=df.loc[df['original_language'].isin(df['original_language'].value_counts().head(10).index)]
```

```python
plt.figure(figsize=(16,8))
plt.subplot(1,2,1)
sns.boxplot(x='original_language',y='revenue',data=language_data)
plt.title('Mean Revenue Per Language')
plt.subplot(1,2,2)
sns.boxplot(x='original_language',y='revenue',data=language_data)
plt.title('Mean log Revenue Per Language')
```
